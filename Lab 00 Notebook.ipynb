```python
'''
    Code cells you can be used to enter python code. This is a 
    multi-line comment. All of your code should be heavily
    commented or accompanied by markdown cells explaining what the 
    code is doing.
'''

# This is a single line comment.

# There are a few add on python packages that will be useful 

# numpy provides a large number of functions for working with
# data. You will use many of its features in calculations and 
# output input.
import numpy as np # you will need to start commands with np.

# create a list of x-values to use in a calculation.
# The following command creates a list of 100 points 
# evenly spaced from 0 to 100.
x = np.linspace(0,10,101)

# What is the 50th element of this list? Keep in mind that 
# Python starts indexing with 0 so the 50th element is 
# element number 49
print(x[49])

'''
    You can use this list to perform more calculations. 
'''
# This code will create a list of y-values that are the 
# sin of 2pi*x. Note that I am using np.pi for 3.1416.
# This constant is defined in the numpy package.
y = np.sin(2*np.pi*x)

# Print out the 50th element of the y-list
print('The 50th element of the y list is: ', y[49])
print('The 50th element of the y list is: %4.3f' %(y[49]))

'''
    You should use a code cell to complete any calculations that
    are needed for a lab. This gives you a record of what numbers 
    were used in a calaculation, allowing you to review the 
    calculation later to double check its accuracy. It is not 
    unusual to make an entry mistake in a calculator that you\
    have no record of.
'''
```

```python
# for graphing we will used matplotlib.plot
import matplotlib.pyplot as plt

'''
    You will often need to make plots of data along with theoretical
    calculations. We will use the package imported above for these.
    There are others out there and you are free to use them if you
    like them more.
'''

# Lets get some data to plot. (use the x and y values from the 
# previous cell.)
plt.plot(x,y)
plt.show()

plt.plot(x,y,'r.')
plt.show()

plt.plot(x,y,'g.-')
plt.show()

plt.plot(x,y,'k--')
plt.ylabel('y data (units)')
plt.xlabel('x data (units)')
plt.title('My graph')
plt.show()

```
